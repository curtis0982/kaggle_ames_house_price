```python
train = pd.read_csv(r"D:\python_code\data\classification\ames_house_prices\train.csv")
test = pd.read_csv(r"D:\python_code\data\classification\ames_house_prices\test.csv")

```

```python
#check the numbers of samples and features
print("The train data size before dropping Id feature is : {} ".format(train.shape))
print("The test data size before dropping Id feature is : {} ".format(test.shape))

#Save the 'Id' column
train_ID = train['Id']
test_ID = test['Id']

#Now drop the  'Id' colum since it's unnecessary for  the prediction process.
train.drop("Id", axis = 1, inplace = True)
test.drop("Id", axis = 1, inplace = True)
```

```python
#檢查outlier
fig, ax = plt.subplots()
ax.scatter(x = train['GrLivArea'], y = train['SalePrice'])
plt.ylabel('SalePrice', fontsize=13)
plt.xlabel('GrLivArea', fontsize=13)
plt.show()
```

```python
#Deleting outliers
train = train.drop(train[(train['GrLivArea']>4000) & (train['SalePrice']<300000)].index)

#Check the graphic again
fig, ax = plt.subplots()
ax.scatter(train['GrLivArea'], train['SalePrice'])
plt.ylabel('SalePrice', fontsize=13)
plt.xlabel('GrLivArea', fontsize=13)
plt.show()
```

```python
#view target dist.
sns.distplot(train['SalePrice'] , fit=norm);

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(train['SalePrice'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Get also the QQ-plot
fig = plt.figure()
res = stats.probplot(train['SalePrice'], plot=plt)
plt.show()
```

```python
#Log-transformation of the target variable
#可看到取log後，變得較為接近常態分配
#We use the numpy fuction log1p which  applies log(1+x) to all elements of the column
train["SalePrice"] = np.log1p(train["SalePrice"])

#Check the new distribution 
sns.distplot(train['SalePrice'] , fit=norm);

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(train['SalePrice'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Get also the QQ-plot
fig = plt.figure()
res = stats.probplot(train['SalePrice'], plot=plt)
plt.show()
```

```python
#Transforming some numerical variables that are really categorical

#MSSubClass=The building class
df['MSSubClass'] = df['MSSubClass'].apply(str)


#Changing OverallCond into a categorical variable
df['OverallCond'] = df['OverallCond'].astype(str)


#Year and month sold are transformed into categorical features.
df['YrSold'] = df['YrSold'].astype(str)
df['MoSold'] = df['MoSold'].astype(str)
```

```python
#We need first to extract the categorial featuers using boolean mask.
# Categorical boolean mask
categorical_feature_mask = df.dtypes==object
# filter categorical columns using mask and turn it into a list
categorical_cols = df.columns[categorical_feature_mask].tolist()
```

```python
categorical_cols
```

```python
#轉換str numeric col全為str
le = LabelEncoder()
for strcol in categorical_cols:
    df[strcol] = le.fit_transform(df[strcol].astype(str))
# apply le on categorical feature columns
df[categorical_cols] = df[categorical_cols].apply(lambda col: le.fit_transform(col))
df[categorical_cols].head(10)
df.fillna(0, inplace=True)

```

```python
#Adding one more important feature
df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']
categorical_feature_mask = df.dtypes==object
```

```python
numeric_feats = df.dtypes[df.dtypes != "object"].index

# Check the skew of all numerical features
skewed_feats = df[numeric_feats].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
print("\nSkew in numerical features: \n")
skewness = pd.DataFrame({'Skew' :skewed_feats})
skewness.head(10)
```

```python
df.shape
```

```python
#Box Cox Transformation of (highly) skewed features

skewness = skewness[abs(skewness) > 0.75]
print("There are {} skewed numerical features to Box Cox transform".format(skewness.shape[0]))

from scipy.special import boxcox1p
skewed_features = skewness.index
lam = 0.15
for feat in skewed_features:
    #df[feat] += 1
    df[feat] = boxcox1p(df[feat], lam)
    
#df[skewed_features] = np.log1p(df[skewed_features])
```

```python
# import OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
# instantiate OneHotEncoder
onehot_enc = OneHotEncoder(categorical_features = categorical_feature_mask, sparse=False ) 
# categorical_features = boolean mask for categorical columns
# sparse = False output an array not sparse matrix
```

```python
# apply OneHotEncoder on categorical feature columns
df_ohe= onehot_enc.fit_transform(df) # It returns an numpy array
```

```python
#4 stage shortcut model 
stack1 = StackingTransformer(estimators=estimators_L1,   # base estimators
                            regression=True,            # regression task (if you need 
                                                        #     classification - set to False)
                            variant='A',                # oof for train set, predict test 
                                                        #     set in each fold and find mean
                            metric=mean_squared_error, # metric: callable
                            n_folds=4,                  # number of folds
                            shuffle=True,               # shuffle the data
                            random_state=0,             # ensure reproducibility
                            verbose=2)                  # print all info
stack2 =stack1
stack3 =stack1
stack4 =stack1
stack5 =stack1

final_estimator =xgb.XGBRegressor(random_state=0, n_jobs=-1, learning_rate=0.1, 
                               n_estimators=100, max_depth=3)
```

```python
#build 4 stage shortcut model
stack1 = stack1.fit(x_train , y_train)
S1_x_train = stack1.transform(x_train )
S1_x_test = stack1.transform(x_test)

S1_x_train = np.concatenate((S1_x_train,x_train),axis=1)
S1_x_test = np.concatenate((S1_x_test,x_test),axis=1)

stack2 = stack2.fit(S1_x_train, y_train)
S2_x_train = stack2.transform(S1_x_train)
S2_x_test = stack2.transform(S1_x_test)

S2_x_train = np.concatenate((S2_x_train,x_train),axis=1)
S2_x_test = np.concatenate((S2_x_test,x_test),axis=1)

stack3 = stack3.fit(S2_x_train, y_train)
S3_x_train = stack3.transform(S2_x_train)
S3_x_test = stack3.transform(S2_x_test)

S3_x_train = np.concatenate((S3_x_train,x_train),axis=1)
S3_x_test = np.concatenate((S3_x_test,x_test),axis=1)

stack4 = stack4.fit(S3_x_train, y_train)
S4_x_train = stack4.transform(S3_x_train)
S4_x_test = stack4.transform(S3_x_test)

S4_x_train = np.concatenate((S4_x_train,x_train),axis=1)
S4_x_test = np.concatenate((S4_x_test,x_test),axis=1)

stack5 = stack5.fit(S4_x_train, y_train)
S5_x_train = stack5.transform(S4_x_train)
S5_x_test = stack5.transform(S4_x_test)

# Initialize 2nd level estimator
final_estimator =xgb.XGBRegressor(random_state=0, n_jobs=-1, learning_rate=0.1, 
                               n_estimators=100, max_depth=3)
# Fit
final_estimator = final_estimator.fit(S5_x_train, y_train)
# Predict
y_pred = final_estimator.predict(S5_x_test)
# Final prediction score
```
